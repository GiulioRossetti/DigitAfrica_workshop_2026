## 1. Key concepts

| Concept | Definition |
|---------|------------|
| **Filter bubble** | Algorithmic curation that limits exposure to diverse viewpoints |
| **Echo chamber** | Social reinforcement of existing beliefs within a group |
| **Epistemic enclave** | A community sharing a distinct and largely self-contained information diet |

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

## 2. A toy example: opinion distribution

The cell below simulates a simple opinion distribution across two polarised groups.

```python
rng = np.random.default_rng(42)

group_a = rng.normal(loc=-0.6, scale=0.15, size=300)  # left-leaning
group_b = rng.normal(loc=+0.6, scale=0.15, size=300)  # right-leaning

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(group_a, bins=30, alpha=0.6, label="Group A", color="steelblue")
ax.hist(group_b, bins=30, alpha=0.6, label="Group B", color="tomato")
ax.axvline(0, color="black", linestyle="--", linewidth=1)
ax.set_xlabel("Opinion score")
ax.set_ylabel("Count")
ax.set_title("Simulated opinion distribution")
ax.legend()
plt.tight_layout()
plt.show()
```
