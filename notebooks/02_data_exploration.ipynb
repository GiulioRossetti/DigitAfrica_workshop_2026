```python
from pathlib import Path

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
sns.set_theme(style="whitegrid")

DATA_RAW = Path("../data/raw")
DATA_PROC = Path("../data/processed")
```

## 1. Load dataset

> **TODO:** Replace the synthetic data below with the actual workshop dataset once it is placed in `data/raw/`.

```python
# --- Synthetic placeholder dataset ---
rng = np.random.default_rng(0)
n = 500
df = pd.DataFrame({
    "user_id": range(n),
    "group": rng.choice(["A", "B"], size=n),
    "activity": rng.poisson(lam=10, size=n),
    "opinion_score": rng.uniform(-1, 1, size=n),
})
print(df.shape)
df.head()
```

## 2. Descriptive statistics

```python
df.describe()
```

```python
df.groupby("group")[["activity", "opinion_score"]].agg(["mean", "std"])
```

## 3. Visualisations

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

sns.histplot(data=df, x="activity", hue="group", bins=20, ax=axes[0])
axes[0].set_title("Activity distribution by group")

sns.histplot(data=df, x="opinion_score", hue="group", bins=20, ax=axes[1])
axes[1].set_title("Opinion score distribution by group")

plt.tight_layout()
plt.show()
```

## 4. Save processed dataset

```python
DATA_PROC.mkdir(parents=True, exist_ok=True)
df.to_csv(DATA_PROC / "users.csv", index=False)
print("Saved to", DATA_PROC / "users.csv")
```
